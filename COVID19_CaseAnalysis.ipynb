```python
cases_df = df.sum()
# storing sum of individual cases for the dataframe
```

```python
cases_df
# subset of dataframe
```

```python
cases_df.drop(['sno', 'state_name', 'state_code', 'total'],inplace=True)
# dropping unnecessary column item
```

```python

# case type and their total sum
cases_df
```

```python
#subsubset1 from cases_df
cases_df1 = cases_df.iloc[:4]
cases_df1
```

```python
cases_df2 = cases_df[4:]
cases_df2
```

```python
# assigning labels and corresponding values for cases for cases_df1
labels= cases_df1.keys()
values=cases_df1.values

# Pie chart
colors = ['#ffcc99','#ff9999','#99ff99','#66b3ff']

#explsion
explode = (0.05,0.05,0.05,1.5)
 
fig1, ax1 = plt.subplots(figsize=(13,8))
ax1.pie(values, colors = colors, labels=labels, autopct='%1.1f%%', startangle=20,pctdistance=0.95, explode = explode)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
# assigning labels and corresponding values for cases for cases_df2
labels= cases_df2.keys()
values=cases_df2.values

# Pie chart
colors = ['#573391','#357c3c','#ef6d6d','#ffe6ab']

#explsion
explode = (0.05,0.05,0.05,1.5)
 
fig1, ax1 = plt.subplots(figsize=(13,8))
ax1.pie(values, colors = colors, labels=labels, autopct='%1.1f%%', startangle=20,pctdistance=0.75, explode = explode)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
# assigning labels and corresponding values for entire cases dataframe
labels= cases_df.keys()
values=cases_df.values

# Pie chart
colors = ['#ffcc99','#ff9999','#99ff99','#66b3ff','#573391','#357C3C','#EF6D6D','#FFE6AB']

#explsion
explode = (0.05,0.05,0.05,0.05,1.5,0.05,0.05,1.5)
 
fig1, ax1 = plt.subplots(figsize=(13,8))
ax1.pie(values, colors = colors, labels=labels, autopct='%1.1f%%', startangle=20,pctdistance=0.55, explode = explode)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
plt.rcParams['figure.figsize']=(13,8)
# giving plots a fixed size of 13*8
```

```python
df_part_byStateName = df[['state_name','active']].groupby(['state_name']).mean().sort_values(by='active')
df_part_byStateName.plot.bar(color='#ff5c8d')
plt.xlabel('State Names')
plt.ylabel('No of Active Cases')
plt.title('Active Cases by States')
plt.show()
```

<h1>Conclusions Drawn as on 21 feb 2022</h1>
<h3>KERELA has the highest number of cases followed by Maharashtra,Karnataka, Rajasthan and so on</h3>
<h3>Dadra AND Nagar Haweli has lowest number of Active cases.</h3>

```python
df_part_byStateName = df[['state_name','death']].groupby(['state_name']).mean().sort_values('death')
df_part_byStateName.plot.bar(color='#b33030')
plt.xlabel('State Names')
plt.ylabel('No of Death Cases')
plt.title('Death Cases by States')
plt.show()
```

<h1>Conclusion Drawn on Death cases as on 21 feb 2022</h1>
<h4>Dadra and Nagar Haweli has least death due to covid 19</h4>
<h4>Maharastra has got the highest number of deaths</h4>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='active', color='teal', data=df)
plt.xlabel('State Names')
plt.ylabel('No of Active Cases')
plt.title('Active Cases by States')

# bar plot visualisation of active cases using seaborn
```

<h3>Once again we find that Kerela has highest no of active cases</h3>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='positive', color='green', data=df)
plt.xlabel('State Names')
plt.ylabel('No of Positive Cases')
plt.title('Positive Cases by States')

# bar plot visualisation of positive cases using seaborn
```

<h3>We conclude that Maharashtra has highest number of positive cases</h3>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='cured', color='teal', data=df)
plt.xlabel('State Names')
plt.ylabel('No of Cured Cases')
plt.title('Cured Cases by States')

# bar plot visualisation of cured cases using seaborn
```

<h3>Maharashtra has cured maximum no of patients of covid 19</h3>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='new_active', color='teal', data=df)
plt.xlabel('State Names')
plt.ylabel('No of New Active Cases')
plt.title('New Active Cases by States')

# bar plot visualisation of new active cases using seaborn
```

<h3>Kerela has got highest number of new active cases all over India</h3>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='new_positive', color='green', data=df)
plt.xlabel('State Names')
plt.ylabel('No of New Positive Cases')
plt.title('New Positive Cases by States')
```

<h3>Maharashtra has got highest number of new positive cases</h3>
<h3>Sikkim and Arunachal Pradesh has got least number of new positive cases</h3>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='new_cured', color='teal', data=df)
plt.xlabel('State Names')
plt.ylabel('No of New Cured Cases')
plt.title('New Cured Cases by States')

# bar plot visualisation of new cured cases using seaborn
```

<h3>Maharashtra shines in new cured cases taking the top positon while least new cured cases are in the states of Chattisgarh,Arunachal Pradesh,Nagaland etc
 but these states are the ones with low positive and low new positive cases as well.</h3>

```python
plt.xticks(rotation=90)
sns.barplot(x='state_name', y='new_death', color='teal', data=df)
plt.xlabel('State Names')
plt.ylabel('No of New Death Cases')
plt.title('New Death Cases by States')

# bar plot visualisation of New Death cases using seaborn
```

<h3>Maharashtra has highest new death cases while least new death cases are in the states of Chattisgarh,Arunachal Pradesh,Nagaland,Mizoram etc.
